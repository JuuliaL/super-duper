```python
# RANDOM WALK SIMULATION

import random

n = 100000 # defining num of steps

x = np.zeros(n)
y = np.zeros(n)

val = random.randint(1, 4)

for i in range(1, n):
    val = random.randint(1, 4)
    if val == 1: # going up
        x[i] = x[i-1] + 1 #dot move x +1 (horizonal)
        y[i] = y[i-1] # y vertical stays the same , needed to start from previus position
    elif val == 2:
        x[i] = x[i-1] - 1 # dot move x -1 (horizontal)
        y[i] = y[i-1] # y vertical stays the same
    elif val == 3:
        x[i] = x[i-1] # x stays the same
        y[i] = y[i-1] + 1 # y move +1 (vertical)
    else:  #val == 4
        x[i] = x[i-1] # x stays the same
        y[i] = y[i-1] - 1 # y move -1 (vertical)

```

```python
import pylab as pl

pl.plot(x, y)

pl.scatter(x[0], y[0], c='green', s=100, label='Start')  # Mark start point
pl.scatter(x[-1], y[-1], c='red', s=100, label='End')  # Mark end point

pl.title(f'Random Walk of n={n} steps.')
pl.xlabel('X axis')
pl.ylabel('Y axis')

pl.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # Increase figure size
plt.plot(x, y, color='blue', lw=2, alpha=0.7)  # Line with transparency
plt.scatter(x[0], y[0], c='green', s=100, label='Start')  # Mark start point
plt.scatter(x[-1], y[-1], c='red', s=100, label='End')  # Mark end point

plt.title('2D Random Walk', fontsize=16)
plt.xlabel('X-Coordinate')
plt.ylabel('Y-Coordinate')
plt.grid(True)
plt.legend()

plt.show()



plt.savefig('2D_rand.png')  #import as png - dont work
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html
```

```python
#TASK
'''Simulate a 1D simple random walk
where at each step can be moving only up or down.
How many steps ⏫  can you actually get?'''

import numpy as np
import matplotlib.pyplot as plt

def random_walk_1d(steps):
    # Initialize the starting position
    position = [0]

    # Generate random steps
    for _ in range(steps):
        step = 1 if np.random.rand() > 0.5 else -1
        position.append(position[-1] + step)

    return position

# Number of steps
n_steps = 1000

# Generate the random walk
walk = random_walk_1d(n_steps)

# Plot the random walk
plt.figure(figsize=(10, 6))
plt.plot(walk, label='1D Random Walk')
plt.title('1D Random Walk Simulation')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.legend()
plt.show()
```

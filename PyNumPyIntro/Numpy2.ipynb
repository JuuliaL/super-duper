```python
import numpy as np
import matplotlib.pyplot as plt

coin = ["H", "T"]
num_throws = 1000
throws = np.random.choice(coin, num_throws)

np.sum(throws == "H")
np.sum(throws == "T")

```

```python
# BIASED COIN TOSS

probabilities = [0.2, 0.8]
p_throws = np.random.choice(coin, num_throws, p=probabilities)
h = np.sum(p_throws == "H")
t = np.sum(p_throws == "T")
print(f"Num of H : {h}")
print(f"Num of T : {t}")
```

```python
# COIN LANDING ON A SIDE

coin_with_side = ["heads", "tails", "side"]
number_of_throws = 10000

p = [0.49, 0.49, 0.02] # last one ise side
throws_with_p = np.random.choice(coin_with_side, number_of_throws, p=p)
np.sum(throws_with_p == "heads")
np.sum(throws_with_p == "tails")
np.sum(throws_with_p == "side")
print(f"heads: {np.sum(throws_with_p == 'heads')}")
print(f"tails: {np.sum(throws_with_p == 'tails')}")
print(f"side: {np.sum(throws_with_p == 'side')}")
```

```python
unique, counts = np.unique(throws_with_p, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
plt.bar(unique, counts)
plt.show()
```

```python
# Sema but longer code
coin = ['heads', 'tails', 'side']
probabilities = [0.49165, 0.49165, 0.0167] #the probability of the coin landing on a side is 0.0167
number_of_throws = 100000
throws = np.random.choice(coin, number_of_throws, p =probabilities)

total_heads_side = np.sum(throws == "heads")
total_tails_side = np.sum(throws == "tails")
total_side = np.sum(throws == "side")

plt.bar(coin, [total_heads_side, total_tails_side, total_side], color="teal")
plt.title("Coin Toss with Side landing")
plt.xlabel("Landing result")
plt.ylabel("Frequency")
plt.show()
```

```python
# HISTOGRAM , Uniform distribution

np.random.randint(1,101) # 100 not included, drop  is last value missing
np.random.randint(1,101, 10) # 10 - last one step

numbers = np.random.randint(1,101, 100000)
plt.hist(numbers, bins=10, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Frequency of rand int from [1, 100]")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
```

```python
coin = ['Heads','Tails'] #one dimentional array == list
```

```python
number_of_throws = 1000
```

```python
#random.choice(a, size=None, replace=True, p=None)

tosses = np.random.choice(coin, size=number_of_throws)
print(f"The result of coin toss simulation for {number_of_throws} throws are: {tosses}")
```

```python
# numpy.sum(a, axis=None, dtype=None, out=None, keepdims=<no value>, initial=<no value>, where=<no value>)

# Number of Heads

number_heads = np.sum(tosses == 'Heads')
```

```python
# Number of Tails

number_tails = np.sum(tosses == 'Tails')
```

```python
# Probabilities
prob_heads = np.sum(number_heads)/number_of_throws
prob_tails = np.sum(number_tails)/number_of_throws

# Uniform distributions - more tosses more probability of going in 50/50 direction

print(f"Probability of Heads: {prob_heads}")
print(f"Probability of Tails: {prob_tails}")
```

### TASK ###

Design a simulation where you toss two coins simultaneously.

Record the possible outcomes (heads or tails for each coin) and run the simulation for multiple trials (e.g., 100, 500, or 1000 tosses).

After collecting the data:

Calculate the probabilities of each outcome (HH, HT, TH, TT).

Visualize the results using a bar chart or pie chart to represent the frequencies and probabilities of each outcome.

Analyze whether the results align with the expected theoretical probabilities.

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_coin_tosses(trials):
    coin_tosses = np.random.choice(['H', 'T'], size=(trials, 2))
    outcomes = [''.join(toss) for toss in coin_tosses]
    outcome_counts = {'HH': 0, 'HT': 0, 'TH': 0, 'TT': 0}
    for outcome in outcomes:
        outcome_counts[outcome] += 1
    return outcome_counts

def calculate_probabilities(outcome_counts, trials):
    return {outcome: count / trials for outcome, count in outcome_counts.items()}

def visualize_results(outcome_counts, probabilities, trials):
    labels = list(outcome_counts.keys())
    frequencies = list(outcome_counts.values())
    probability_values = list(probabilities.values())

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.bar(labels, frequencies, color=['blue', 'orange', 'green', 'red'])
    plt.xlabel('Outcomes')
    plt.ylabel('Frequencies')
    plt.title(f'Frequencies of Coin Toss Outcomes ({trials} trials)')

    plt.subplot(1, 2, 2)
    plt.pie(probability_values, labels=labels, autopct='%1.1f%%', colors=['blue', 'orange', 'green', 'red'])
    plt.title(f'Probabilities of Coin Toss Outcomes ({trials} trials)')

    plt.tight_layout()
    plt.show()

# Expected theoretical probabilities
theoretical_probabilities = {'HH': 0.25, 'HT': 0.25, 'TH': 0.25, 'TT': 0.25}

# Run simulations for different numbers of trials
for trials in [100, 500, 1000]:
    outcome_counts = simulate_coin_tosses(trials)
    probabilities = calculate_probabilities(outcome_counts, trials)
    visualize_results(outcome_counts, probabilities, trials)

    print(f"Results for {trials} trials:")
    for outcome in theoretical_probabilities:
        print(f"{outcome}: Simulated Probability = {probabilities[outcome]:.4f}, Theoretical Probability = {theoretical_probabilities[outcome]:.4f}")
    print("\n")
```

# UNSUPERVISED TRAINING EXAMPLE (CLUSTERING)
https://scikit-learn.org/dev/api/sklearn.datasets.html

```python
from sklearn.datasets import make_blobs # create sytethic dataset
```

```python
X, y = make_blobs(n_samples=1000, centers=6, cluster_std=0.80) # std bigger more disperce is blobs
```

```python
X[:, 0] # x coordinate (2 arrays filled with values)
X[:, 1] # y coordinate
y # the cluster of the point
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c='gray', edgecolors='k') # c=y  colorful
```

```python
from sklearn.cluster import KMeans #import model
```

```python
kmeans = KMeans(n_clusters=3) # initialize, assume we dont know we have 6 clusters
```

```python
kmeans.fit(X) # train model
```

```python
cluster_labels = kmeans.predict(X) # predict cluster, as we defined n_clusters=2
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, edgecolors='k') # color to labels
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], edgecolors='k', s=300, marker='*', c='red') # centers
```

```python
kmeans.cluster_centers_ # center on each cluster
kmeans.cluster_centers_[:, 0] # x coordinate
kmeans.cluster_centers_[:, 1] # y coordinate
```

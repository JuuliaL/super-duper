```python
import re # regular expressions module
```

```python
text = """
Alice: 555-123-4567
Bob: 333-987-6543
Charlie: 777-555-88
"""
# """ for multiline
```

```python
print(text)
# correct format is 3-3-4

pattern = r"\d{3}-\d{3}-\d{4}"   # r'\b\d{3}-\d{3}-\d{4}\b'  - \b for word boundaries

matches = re.findall(pattern, text)

print(matches)
```

```python
# re.search - first occurance of a pattern
# re.findall - finds all occurances
# re.IGNORECASE - pattern matching is case-sensitive
```

```python
# Dates pattern matching

dates = """
My dates are: 05-10-2024 and 2022-10-05
"""

pattern2 = r"\d{2}-\d{2}-\d{4}"

matches2 = re.findall(pattern2, dates)

print(matches2)
```

### E-mail addresses matching

```python
email = """
Emails: liza@example.com, roberta@gmail.com, info123@company.net
"""
# requested output ['liza@example.com', 'roberta@gmail.com', 'info123@company.net]

pattern3 = r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b"

matches3 = re.findall(pattern3, email)

print(matches3)
```
